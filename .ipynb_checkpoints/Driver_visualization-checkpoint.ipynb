We now run simulations batches when the boarder is set on fire with different densities.

```python
border_data = data_border(20,30, (.1,.2,.3,.4,.45,.5,.55,.6,.7,.8,.9) )
border_data_labels = ['border']*len(border_data)
```

```python
df = pd.DataFrame(border_data)
#df = df.transpose()
df.columns = ['X', 'Y']
#df
```

We examine the plots of the data and the distributions.

```python
sns.pairplot(df, hue="X")
```

Now for some box-plots

```python
sns.set(style="darkgrid")
g = sns.jointplot("X", "Y", data=df, kind="reg",
                  xlim=(-.1, 1), ylim=(-.1, 1), color="b", height=5)
```

Data looks non normal for all values of density, and linear regression seems like a poor fit! We will try to fit a sigmoid/logistic regression.

```python
X = df['X'].values
y = df['Y'].values
#X = X.reshape(-1, 1)
#y = y.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=25)
```

Fit a sigmoid function to the data.

```python
def func(x, a, b):
    return 1/(1+b*np.exp(a*x)) 

# Fit the parameters of the function with noisy data:
params, pcov = curve_fit(func, X , y)
yfit = func(X, params[0], params[1])

# Plot the outcome:
fig = plt.figure
plt.plot(X, yfit, "k-")
plt.plot(X, y, "bx")
plt.savefig('fit.png')
```

```python
params
```

The plot the residuals suggests heteroscedasticity. 

```python
residual = y - yfit
figRes = plt.figure
plt.plot(X, residual, "bx")
plt.savefig('figRes.png')
```
